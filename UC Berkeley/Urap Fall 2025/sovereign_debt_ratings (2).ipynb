```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_credit_rating"
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0 Safari/537.36"
}
```

```python
# Fetch and parse page
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# All 'wikitable' tables
tables = soup.find_all("table")
#tables = soup.select("table.wikitable")
print(f"Found {len(tables)} tables on the page.")
```

```python
# known rating labels 
agency_labels = [
    "Standard & Poor's",
    "Fitch",
    "Moody's",
    "DBRS Morningstar",
    "Scope Ratings",
    "JCR",
    "China Chengxin",
    "CareEdge Global Ratings",
    "Unrated"
]

# use lxml
df = [] 
for table in tables:
    df.append(pd.read_html(str(table))[0])

# dict of DataFrames
ratings_dict = {}
for agency, i in zip(agency_labels, range(0,9)):
    ratings_dict[agency] = df[i]
    print(f"{agency}: {len(df)} rows")

# S&P 
ratings_dict["Standard & Poor's"].head()
```

```python
from IPython.display import FileLink, display
#saving as csv
for agency, dataframe in ratings_dict.items():
    filename = agency.replace(" ", "_").replace("'", "") + ".csv"
    path = os.path.join("credit_ratings_csv", filename)
    dataframe.to_csv(path, index=False)
    display(FileLink(path)) 
```
